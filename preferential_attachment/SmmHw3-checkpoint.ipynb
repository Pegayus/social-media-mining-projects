Problem 6: Preferential Attachment for a graph with 10 inital node and expected degree of 7, adding 40 more nodes.

```python
import networkx as nx
import matplotlib.pyplot as plt
m0=10
m=7
t=40  
```

Generating the initial graph:

```python
G=nx.expected_degree_graph([5]*m0,None,False)
edg=[0]*m0
for j in range(0,m0):
    edg[j]=G.degree(j)
total_edge=sum(edg)
p=[0]*m0
for j in range(0,m0):
    p[j]=G.degree(j)/total_edge
```

Adding nodes and edges:

```python
for i in range(1,t+1):
    G.add_node((m0-1)+i)
    while G.degree(m0-1+i)<m:
        max_index=p.index(max(p))
        G.add_edge(max_index,m0-1+i)
        edg[max_index]=edg[max_index]+1
        total_edge=total_edge+1
        p[max_index]=G.degree(max_index)/total_edge
```

Plotting:

```python
nx.draw(G)
plt.show()
```
